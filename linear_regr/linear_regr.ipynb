```python
#Linear Regression
#####[using salary dataset to train the model and then evaluating the model by finding the accuracy score by comparing the predicted data(salary)]
#load the dataset
import pandas as pd
data=pd.read_csv("Salary_Data ML-2.csv")
print(data)

#data cleaning
#data.isna().sum()   #finds the count of missing valuws un the dataset
data.dropna(inplace=True)      #removes that missing values from dataset
data.isna().sum()

#data preprocessing
from sklearn.preprocessing import LabelEncoder
for i in data.columns:
    if data[i].dtype=='object':
        le=LabelEncoder()
        data[i]=le.fit_transform(data[i])

print(data)

#Training-Testing-Splitting:-
#spliting data into x(independent features) and y(dependent feature)
x=data.iloc[:,:-1]  #input features
y=data.iloc[:,-1]   #target/output
#splitting into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=52)
#learn and train using linear regression algorithm
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train) #train model on provided data‚Üê yahi step model ko sikhata hai
yp=model.predict(x_test) #testing model to predict the values
print(yp)

from sklearn.metrics import r2_score
r2_score(y_test,yp)         #to test the accuracy of predicted values compared to actual values


import matplotlib.pyplot as plt

plt.scatter(y_test, yp)
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary")
plt.show()
```

```python
#Linear Regression

########[using car data for car price prediction ]

#load the dataset
import pandas as pd
car_data=pd.read_csv("car data.csv")
print(car_data)

#inspecting the first 5 rows of the dataframe
car_data.head()


```

```python
#getting some information about the dataset
car_data.info()
```

```python
#getting the number of missing values
car_data.isna().sum()
```

```python
#Checking the distribution of categorical data
print(car_data.Fuel_Type.value_counts())
print(car_data.Seller_Type.value_counts())
print(car_data.Transmission.value_counts())
```

```python
#Data Preprocessing :- Encoding the categorical data

from sklearn.preprocessing import LabelEncoder
for i in data.columns:
    if car_data[i].dtype=='object':
        le=LabelEncoder()
        car_data[i]=le.fit_transform(car_data[i])

print(car_data)

car_data.head()
```

```python
#Training-Testing-Splitting:-

#spliting the data into x(independent features) and y(dependent feature)
x = car_data.drop(['Car_Name','Selling_Price'],axis=1)
y = car_data['Selling_Price']

#splitting into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=2)

#learn and train our model using linear regression algorithm
from sklearn.linear_model import LinearRegression  #Loading the linear regression model
lin_reg_model=LinearRegression()

lin_reg_model.fit(x_train , y_train)    #train model on provided data‚Üê yahi step model ko sikhata hai

#prediction on training data
training_data_prediction = lin_reg_model.predict(x_train)  #prediction on training data

#Model evaluation
from sklearn import metrics
#R squared error
error_score = metrics.r2_score(y_train,training_data_prediction )
print("R squared error:",error_score)

#Visualize the actual prices and predicted prices
import matplotlib.pyplot as plt
import seaborn as sns

plt.scatter(y_train , training_data_prediction )
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted prices of car")
plt.show()
```

```python
#prediction on test data
test_data_prediction = lin_reg_model.predict(x_test)  #prediction on test data
#Model evaluation
from sklearn import metrics
#R squared error
error_score = metrics.r2_score(y_test,test_data_prediction )
print("R squared error:",error_score)

#Visualize the actual prices and predicted prices
import matplotlib.pyplot as plt
import seaborn as sns

plt.scatter(y_train , training_data_prediction )
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted prices of car")
plt.show()
```
